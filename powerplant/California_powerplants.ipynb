```python
import pandas as pd
```

```python
df = pd.read_csv("California_Power_Plants.csv")
```

```python
df.info()
```

```python
df.sort_values(by="Capacity_Latest", ascending=False).head()
```

```python
df.drop(["RPS_Eligible"], axis=1, inplace=True)
df.describe()
```

```python
dd = df.groupby("PriEnergySource").sum("Capacity_Latest").sort_values(by="Capacity_Latest", ascending=False)
```

```python
dd = dd[dd["Capacity_Latest"] > 1000]
dd
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.pie(dd["Capacity_Latest"], labels= dd.index, autopct='%1.1f%%')
ax.set_title("California Energy Capacity")
```

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
df.columns
```

```python
geometry = [Point(xy) for xy in zip(df['Lon_WGS84'], df['Lat_WGS84'])]
gdf = gpd.GeoDataFrame(df, geometry=geometry)
print(gdf)
```

```python
geopandas.datasets.available
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(15,15)), marker='o')
```
