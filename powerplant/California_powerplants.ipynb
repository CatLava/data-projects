```python
df['geometry'] = [Point(xy) for xy in zip(df['Lon_WGS84'], df['Lat_WGS84'])]
geof = df.filter(['PriEnergySource', 'Capacity_Latest', 'geometry', 'Lon_WGS84', 'Lat_WGS84'], axis=1)
gdf = gpd.GeoDataFrame(geometry=geometry)
print(gdf)
```

```python
gf = geof[geof["Capacity_Latest"] > 100]
gf['PriEnergySource'].unique()
colors = {
    'NG' : 'blue',
    'WND': 'red',
    'SUN': 'yellow',
    'WAT': 'green',
    'NUC': 'purple',
    'unknown' : 'white',
    'GEO': 'brown',
    'JF': 'black',
    'NG, DFO' : 'blue',
    'NG, OG': 'blue',
    'BIT' :'orange'
}
gf.dropna()
gf['PriEnergySource'] = gf['PriEnergySource'].fillna("unknown")
gf['color'] = gf.apply(lambda row: colors[row['PriEnergySource']], axis=1)
gf
```

```python
newdgf = gpd.GeoDataFrame(geometry=gf['geometry'])
print(newdgf)
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(15,15)), marker='o', color='red', markersize=15)
```
